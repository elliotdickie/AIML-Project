```python
import csv

#function for reading data from pgm to list of integers
def readpgm(name):
    with open(name) as f:
         lines = f.readlines()

    for l in range(2):
        lines.remove(lines[l])

    # Converts data to a list of integers
    data = []
    for line in lines[1:]:
        data.extend([int(c) for c in line.split()])

    return data

#converts list of integers to 18 by 18 csv
def createCSV(name,data):
    csvfile = open(name+'.csv','w', newline = '', encoding = 'utf-8')

    writer = csv.writer(csvfile)
    for l in range(18):
        row = []
        for i in range(18):
            row.append(data[l*18+i])
        writer.writerow(row)

    csvfile.close()

#calls the two functions for every letter
chars = ['a','b','c','d','e','f','g','h','i','j']
for char in chars:
    for x in range(1,9):
        name = 'images/40431475_'+char+'_'+str(x)
        data = readpgm(name + '.pgm')
        createCSV(name,data)

#calls the two functions for every non-letter
emoticons = ['sad','smiley','xclaim']
for emoticon in emoticons:
    for x in range(1,21):
        name = 'images/40431475_'+emoticon+'_'+str(x)
        data = readpgm(name + '.pgm')
        createCSV(name,data)
```
